```python
# type B_n
n=8;
a = matrix(QQ,2*n+1);
J1=[1,2,3,7];J2=[4,5,6];
# a[0,2*n]=1;a[n,0]=-1; #n in J1
a[0,n]=1;a[2*n,0]=-1; #n in J2
for i in J1:
    a[i,i+1]=1;
    a[i+1+n,i+n]=-1;
for i in J2:
    a[i+1,i]=1;
    a[i+n,i+1+n]=-1;
show(a.jordan_form());
b=copy(a);
for i in range(2*n+1):
    for j in range (2*n+1):
        b[i,j]=abs(b[i,j]);
# show(b.jordan_form());
```

```python
# type D_n
n=5;
a = matrix(QQ,2*n);
J1=[1,2,4];J2=[3];
# a[n-2,2*n-1]=1;a[n-1,2*n-2]=-1; #n in J1
a[2*n-2,n-1]=1;a[2*n-1,n-2]=-1; #n in J2
for i in J1:
    a[i-1,i]=1;
    a[i+n,i-1+n]=-1;
for i in J2:
    a[i,i-1]=1;
    a[i-1+n,i+n]=-1;
show(a.jordan_form());
b=copy(a);
for i in range(2*n):
    for j in range (2*n):
        b[i,j]=abs(b[i,j]);
# show(b.jordan_form());
c=copy(a); #relabeling rows/columns to 1,..,n,2n,..,n+1
for k in range(floor(n/2)):
    for j in range(n*2):
        temp=c[n+k,j];
        c[n+k,j]=c[2*n-1-k,j];
        c[2*n-1-k,j]=temp; #switch rows
for k in range(floor(n/2)):
    for i in range(n*2):
        temp=c[i,n+k];
        c[i,n+k]=c[i,2*n-1-k];
        c[i,2*n-1-k]=temp; #switch columns
show(c*c);
        
   
```
